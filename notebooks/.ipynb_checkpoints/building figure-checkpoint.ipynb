```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

data = [[100 , 100, 0],
        [77, 82.7, 12],
        [74.9, 71.7, 24], 
        [67.6, 61.4, 32]]
data = np.array(data)
std = np.std(data[:, :2], axis=1)
mean = np.mean(data[:, :2], axis=1)

# plt.figure(dpi=150)
# plt.scatter(data[:, -1], mean)
plt.errorbar(data[:, -1], mean, yerr=std, fmt="o")
plt.plot(data[:, -1], mean, ls='--', c='#1f77b4')
plt.fill_between(data[:, -1], mean-std, mean+std, color='gray', alpha=0.1)

plt.xlabel("$N_{C}$",fontsize=12)
plt.ylabel(r"$\frac{Abs_m}{Abs_{P-O}}$", fontsize=12)
plt.title("Octanol-Water Partitioning of Aliphatic Series", fontsize=13)
plt.axhline(100, ls='--', c='r')

plt.scatter(data[:, -1], 100*np.ones(mean.shape), c='r')
plt.plot(data[:, -1], 100*np.ones(mean.shape), ls='--', c='r')
plt.savefig("error_bar_part.png", dpi=300)
plt.show()
```
